```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12, 12))
x1 = np.linspace(0,1.5,100)
y = x1+1/2
x2 = np.linspace(1.5,3.5,100)
z = -x2+7/2
min_y_z = []
for i in range(len(y)):
    min_y_z.append(y[i])
for i in range(len(z)):
    min_y_z.append(z[i])
min_y_z=np.array(min_y_z)
plt.xlim(0,4)
plt.ylim(0,3.5)
ax.plot(x1, y, '-r', label='$x_1-x_2 = -1/2$')
ax.plot(x2, z, '-y', label='$x_1+x_2 = 7/2$')
ax.plot([0,0,0,1,1,1,2,2,2,3,3,3],[0,1,2,0,1,2,0,1,2,0,1,2],'o')
ax.annotate("", xy=(0, 3.5), xytext=(0, 0),arrowprops=dict(facecolor='black', shrink=1))
ax.annotate("", xy=(4, 0), xytext=(0, 0),arrowprops=dict(facecolor='black', shrink=1))
ax.annotate("Optiaml solution for\n the LP relaxation", fontsize=18, xy=(1.5, 2), xytext=(2.6, 1.7),arrowprops=dict(width=1,headwidth=10))
ax.annotate("The rounded solutions\n are not feasible", fontsize=18, xy=(1.5, 2.7), xytext=(1.5, 3.2),arrowprops=dict(width=0,headwidth=0))
ax.annotate("", xy=(1.01, 2.01), xytext=(1.5, 2.7),arrowprops=dict(width=1,headwidth=10))
ax.annotate("", xy=(2.01, 2.01), xytext=(1.5, 2.7),arrowprops=dict(width=1,headwidth=10))
ax.set_xlabel('$x_1$',fontsize=18)
ax.set_ylabel('$x_2$',fontsize=18)
ax.set_title('Graph',fontsize=18)
ax.text(1.3,2.1,'(3/2,2)',fontsize=18)
ax.fill_between(np.hstack((x1,x2)),0,min_y_z,facecolor="green", alpha=0.2, label="Feasible Region")
ax.legend(loc='upper right',fontsize=18)
ax.grid()
```
