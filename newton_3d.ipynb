```python
import numpy as np
```

```python
def functionValue(x):
    f=[]
    f.append([3*x[0,0]-np.cos(x[1,0]*x[2,0])-0.5])
    f.append([x[0,0]**2 - 625 * x[1,0]**2 - 0.25])
    f.append([np.exp(-x[0,0]*x[1,0])+ 20 * x[2,0] + ((10*np.pi-3)/3)])
    f = np.array(f)
    return f
```

```python
def functionJacobian(x):
    Df = []
    Df.append([3                              , x[2,0]*np.sin(x[1,0]*x[2,0])  , x[1,0]*np.sin(x[1,0]*x[2,0]) ])
    Df.append([2*x[0,0]                       , -1250*x[1,0]                  , 0                            ])
    Df.append([-x[1,0]*np.exp(-x[0,0]*x[1,0]) ,-x[0,0]*np.exp(-x[0,0]*x[1,0]) , 20                           ])
    Df = np.array(Df)
    return Df
```

```python
def JacobianValue(x):
    value = 3*-1250*x[1,0]*20 + x[1,0]*np.sin(x[1,0]*x[2,0])*2*x[0,0]*-x[0,0]*np.exp(-x[0,0]*x[1,0]) - x[1,0]*np.sin(x[1,0]*x[2,0])*-1250*x[1,0]*-x[1,0]*np.exp(-x[0,0]*x[1,0])-20*x[2,0]*np.sin(x[1,0]*x[2,0])*2*x[0,0]
    return value
```

```python
x=[[1],[1],[1]]
x = np.array(x)
k = 1

print("Initial x1,x2")
print(x)
print()

while JacobianValue(x) != 0 and k <= 100:
    Dx = -np.linalg.inv(functionJacobian(x)).dot(functionValue(x))
    x = x + Dx
    print("Time: ",end = ' ')
    print(k)
    print(x)
    print()
    print(JacobianValue(x))
    print()
    k = k+1
```
