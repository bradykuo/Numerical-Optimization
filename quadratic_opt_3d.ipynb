```python
import numpy as np

H = np.array([[10,-18,2],[-18,40,-1],[2,-1,3]])
c = np.array([[12],[-47],[-8]])

x_origin = []
x_origin.append([[0],[0],[0]])
x_origin.append([[15.09],[7.66],[-6.56]])
x_origin.append([[11.77], [6.42], [-4.28]])
x_origin.append([[4.46], [2.25], [1.85]])
```

```python
def functionValue(x):
    y = 1/2* np.transpose(x).dot(H).dot(x) + np.transpose(c).dot(x)
    return y[0][0]

def functionGradient(x):
    y = np.matmul(H,x) + c
    return y

def stepSize(x, d):

    a = -float((10*x[0,0]*d[0,0] + 40 * x[1,0]*d[1,0] + 3*x[2,0]*d[2,0] - 18*x[0,0]*d[1,0] - 18 * x[1,0]*d[0,0] +
         2*x[0,0]* d[2,0] + 2* x[2,0]*d[0,0] - x[1,0]*d[2,0] -x[2,0]*d[1,0] + 12 * d[0,0] - 47* d[1,0] - 8* d[2,0]))/float(
             2*(5*d[0,0]**2 + 20*d[1,0]**2 + 1.5*d[2,0]**2 - 18*d[0,0]*d[1,0] - 2*d[0,0]*d[2,0] - d[1,0]*d[2,0]))

    return a
```

```python
for w in x_origin:
    x = [np.array(w)]
    dk = [-functionGradient(x[0])]
    k = 0

    while np.linalg.norm(dk[k])>1e-6 and k <= 10000:
        x.append(x[k] + stepSize(x[k], dk[k])*dk[k])
        k = k+1
        dk.append(-functionGradient(x[k]))

    print("The origin is (%.6f,%.6f,%.6f)" % (x[0][0], x[0][1], x[0][2]))
    print("The value of start is %.6f\n" % functionValue(x[0]))
    print("An optimal solution is given by (%.6f,%.6f,%.6f)" % (x[-1][0], x[-1][1], x[-1][2]))
    print("The optimal value is reported as %.6f" % functionValue(x[-1]))
    print("(total take", k, 'step)\n\n')
```
