```python
import numpy as np
import random


def determineStepSize(x,d,r):
    n=len(d)
    stepSize=float("inf")
    for i in range(n):
        if d[i] < 0 and stepSize>-x[i]/d[i]:
            stepSize=-x[i]/d[i]
            
    stepSize=min([1,r*stepSize])
    return stepSize

def IPM(A,b,c,x0):  
    y0 = [[random.randint(1,100)]]
    s0 = [[random.randint(1,100)],[random.randint(1,100)],[random.randint(1,100)]]
    mu=[10]
    r=0.6
    mu_0=10
    epsilon=1e-6
    e=np.ones((3,1))
    k=0

    while mu[k]>epsilon*mu_0:
        #mu[k]=mu_0*(1/10)**k
        x0=np.array(x0)
        y0=np.array(y0)
        s0=np.array(s0)
        solution=np.vstack((x0,y0,s0))

        X=np.diag(x0[:,0])
        S=np.diag(s0[:,0])

        r1=b-A.dot(x0)
        r2=c-A.transpose().dot(y0)-s0
        r3=mu[k]*e-X.dot(S).dot(e)
        residual=np.vstack((r1,r2,r3))

        A1=np.hstack((A,np.zeros((1,1)),np.zeros((1,3))))
        A2=np.hstack((np.zeros((3,3)),A.transpose(),np.identity(3)))
        A3=np.hstack((S,np.zeros((3,1)),X))
        barA=np.vstack((A1,A2,A3))

        NewtonDirection=np.linalg.inv(barA).dot(residual)
        delta_x=NewtonDirection[0:3,0].reshape(3,1)
        delta_y=NewtonDirection[3:4,0].reshape(1,1)
        delta_s=NewtonDirection[4:7,0].reshape(3,1)

        stepSizeX=determineStepSize(x0,delta_x,r)
        stepSizeS=determineStepSize(s0,delta_s,r)
        x0=x0+stepSizeX*delta_x
        y0=y0+stepSizeS*delta_y
        s0=s0+stepSizeS*delta_s
        #mu[k+1]=(1/10)*x0.transpose().dot(s0)/4#complementary condition
        mu.append((1/10)*x0.transpose().dot(s0)/3)
        print("mu is ",mu[k+1])
        #print("stepSizeX is ",stepSizeX)
        #print("stepSizeS is ",stepSizeS)
        print('k=',k)
        print(x0)
        k=k+1
    print("optimal value =", c.transpose().dot(x0))

```
