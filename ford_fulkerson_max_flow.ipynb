```python
class Network:

    def __init__(self,network): #object's constructor
        self.network = network  #object's attribute
        self.nodeNum = len(network)

    def BFS(self, source, sink, parent): #object's method
        visited = [False]*(self.nodeNum) #用來記錄是否被拜訪過
        temp = []
        temp.append(source)
        visited[source] = True

        while temp:
            first = temp.pop(0)

            for second, capacity in enumerate(network[first]): #enumerate-> [(0,capacity from i to 0),(1,capacity from i to 1),...]
                if visited[second] == False and capacity > 0:  #若有capacity且沒被拜訪過代表可相連
                    temp.append(second) #將可以相連的node存到這個list，之後再從temp[0]繼續找可以相連的
                    visited[second] = True
                    parent[second] = first #parent list 用來記錄往前的路徑
                    if second == sink:
                        return True #有成功找到BFS
        return False #無法找到BFS(沒有可往後送的capacity)

    def FordFulkerson(self, source, sink):
        parent = [-1]*(self.nodeNum)
        maxFlow = 0

        while self.BFS(source, sink, parent):
            pushFlow = float("Inf")

            a = sink
            while(a != source): #從終點往前找路徑，找出capacity最小(可推送流量)
                pushFlow = min(pushFlow, self.network[parent[a]][a])
                a = parent[a] #往前找路徑

            maxFlow += pushFlow #每條路徑的最小可推送流量相加就是整個網路的最大流量

            b = sink
            while(b != source):#從終點往前找路徑，調整各node之間的capacity
                self.network[parent[b]][b] -= pushFlow #往終點送的capacity會減少
                self.network[b][parent[b]] += pushFlow #可反悔capacity(反方向)會增加
                b = parent[b]
        return maxFlow
```
