```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#prob_2

#define plt object
fig, ax2 = plt.subplots(figsize = (8, 8))

#axis
plt.xlim(-0.5, 3.5)
plt.ylim(-0.5, 3.5)
ax2.annotate("", xy=(0, 3.3), xytext=(0, 0),arrowprops=dict(width=0.001,headwidth=7, color = 'black'))
ax2.annotate("", xy=(3.3, 0), xytext=(0, 0),arrowprops=dict(width=0.001,headwidth=7, color = 'black'))
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
#int dots
ax2.plot([0, 0, 1, 2, 0, 1, 2, 0, 1, 2, 3], [3, 2, 2, 2, 1, 1, 1, 0, 0, 0, 0], 'o', markersize=9, color = 'black')

#constrain limits
x1 = np.linspace(0, 2)
y1 = 2 - x1/10
ax2.plot(x1, y1, 'black', linewidth = '3')
ax2.plot([0 for i in range(50)], np.linspace(0, 2),'black', linewidth = '3')
ax2.plot(np.linspace(2, 2), np.linspace(0, 1.8), 'black', linewidth = '3')

#optimal function (IP)
x2 = np.linspace(-0.3, 2.7)
y2 = 2 - x2/5
ax2.plot(x2, y2, '--', color = 'black', label = '$Z* = 10 = x_1 + 5x_2$')

#feasible region
ax2.fill_between(x1, 0, y1, color = 'lightgrey')

# optimal IP solution
ax2.annotate("", xy=(0.05, 2.1), xytext=(0.5, 3.1),arrowprops=dict(width=0.0001,headwidth=7, color = 'black'))
ax2.text(0.6, 3.1,'Optimal IP solution', fontsize = 12)

# optimal solution for the LP realxation
ax2.annotate("", xy=(2.1, 1.9), xytext=(2.5, 2.2),arrowprops=dict(width=0.0001,headwidth=7, color = 'black'))
ax2.text(2.3, 2.3,'Optimal solution for\n the LP relaxation', fontsize = 12)

# Rounded solution
ax2.annotate("", xy=(2.1, 1), xytext=(2.5, 0.8),arrowprops=dict(width=0.0001,headwidth=7, color = 'black'))
ax2.text(2.3, 0.7,'Rounded solution', fontsize = 12)

ax2.legend(loc='upper right',fontsize=12)
#ax2.grid()
```
